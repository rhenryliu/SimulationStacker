```python
def load_measurements_npz(path):
    """Load NPZ saved by save_measurements_npz back into nested dict."""
    z = np.load(path)
    out = {}
    for k in z.files:
        name, key = k.split("/", 1)
        out.setdefault(name, {})[key] = z[k]
    return out

measurements = load_measurements_npz(filePath + fileName)
```
